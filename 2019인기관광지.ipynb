```python
import pandas as pd
```

```python
tour19 = pd.read_csv("20240109005451_인기관광지_전체.csv", encoding="CP949")
```

```python
tour19.head()
```

```python
tour19.columns
```

```python
tour19.rename(columns={'Unnamed: 4':'방문자'}, inplace=True)
tour19.head()
```

```python
tour19_addr = tour19["주소"]
```

```python
addr_lst = []
tour19_addr_lst= []
```

```python
# 첫 번째, 두 번째 어절만 가지고 와야 함
# 그러려면 먼저 어절 단위로 쪼개야 하고, 그걸 위해서는 split()을 사용해야 함
# 그러나 현재는 series 형태
# 리스트를 사용해서 "구"만 뽑아내기
```

```python
for i in range(len(tour19_addr)):
    tour19_addr_lst.append(tour19_addr[i].split()[1])


len(tour19_addr_lst)
```

```python
tour19["주소"] = tour19_addr_lst
```

```python
tour19_df = tour19[['관광지명', '주소', '방문자']]
```

```python
tour19_df.head()
```

```python
#행정구별 총방문자수
tour19_addr_total = tour19_df.groupby('주소')[['방문자']].sum()
tour19_addr_total#.sort_values(by="방문자", ascending=False)
```

```python
#서울 평균 방문자수
tour19_addr_mean = tour19_df.groupby('주소')[['방문자']].mean()
tour19_addr_mean
```
