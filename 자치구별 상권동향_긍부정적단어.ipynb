```python
import pandas as pd
```

```python
place=pd.read_csv("C:/Users/YujinCho/OneDrive/DAS/data/프로젝트1/자치구별상권동향.csv")
place
```

```python
place.columns=['구분번호','기준년도','시군구명','시군구코드','긍정적단어','긍정적단어 횟수','부정적단어','부정적단어 횟수','데이터로드일자']
place.info()
```

```python
new_row = pd.DataFrame([[1,2022,'마포구',11442,'좋아하다',28,'불량식품',2,'2023-04-19']], columns = place.columns)
prac_place = pd.concat([place.iloc[:0], new_row, place.iloc[0:]], ignore_index=True)
```

```python
place = prac_place.copy()
place
```

```python
# 부정적단어 횟수의 내림차순
place.groupby('시군구명').mean().sort_values(by=['부정적단어 횟수'], axis=0,ascending=False)
```

```python
# 긍정적단어 횟수의 내림차순
place.groupby('시군구명').mean().sort_values(by=['긍정적단어 횟수'], axis=0,ascending=False)
```

```python
negative_words_by_region = place.groupby('시군구명')['부정적단어'].agg(list).reset_index()
negative_words_by_region
```

```python
ratio_ver=place.groupby('시군구명').mean().sort_values(by=['부정적단어 횟수'], axis=0,ascending=False)
ratio_ver['부정비율']=ratio_ver['부정적단어 횟수']/ratio_ver['긍정적단어 횟수']
ratio_ver.sort_values(by=['부정비율'],axis=0,ascending=False)
```
