```python
popreg_23=pd.read_csv("C:/Users/YujinCho/OneDrive/DAS/data/프로젝트1/20240109005532_서울특별시_202301-202311_데이터랩_다운로드/20240109005532_인기관광지_전체.csv", encoding='cp949')
popreg_23
```

```python
popreg_23.rename(columns={'Unnamed: 4':'방문객 수'}, inplace=True) #바뀐 열 이름 적용
popreg_23
```

```python
#총 방문객 수, 평균 방문객 수
a = [ popreg_23['방문객 수'].sum(), popreg_23['방문객 수'].mean() ]
print(f"총 방문객 수: ,{a[0]}")
print(f"평균 방문객 수:. {a[1]}")
```

```python
# 행정구역 나누기
popreg_23['행정구역'] = popreg_23['주소'].apply(lambda x: x.split('구')[0] + '구' if '구' in x else x)
popreg_23
```

```python
# 행정구역 잘 나뉘었는지 체크 -> '서울 구'를 '서울 구로구'로 대체
print(popreg_23['행정구역'].unique())
print('-------------------------------------------')
lst=popreg_23[popreg_23['행정구역']=='서울 구'].index
print(lst) 
print('-------------------------------------------')
popreg_23.replace('서울 구', '서울 구로구', inplace=True)
print(popreg_23['행정구역'].unique())
```

```python
# 행정구역별 총 방문객 수, 평균 방문객 수
popreg_23.groupby('행정구역').agg(['sum', 'count', 'mean']).round(2)
```
